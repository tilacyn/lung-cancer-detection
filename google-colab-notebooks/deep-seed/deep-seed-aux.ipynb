```
import numpy as np
os.chdir(luna_path)
test_idcs = np.load('luna_test.npy')
```

```
from zipfile import ZipFile

os.chdir(opjoin(base_path, 'data'))

def zipdir(path, ziph):
    # ziph is zipfile handle
    for root, dirs, files in os.walk(path):
        for file in files:
            if file[:3] not in test_idcs:
              continue
            print(file)
            print('processing file ', file)
            ziph.write(os.path.join(root, file))
        for subd in dirs:
            print('processing dir ', subd)
            ziph.write(os.path.join(root, subd))

with ZipFile('preprocess-luna-test.zip', 'w') as myzip:
    zipdir(opjoin(base_path, 'data', 'preprocess-result-path'), myzip)
```
